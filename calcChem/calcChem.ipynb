```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset_id = pd. read_csv("p_id.txt", sep=" ", header=None, names=["V", "p"])
dataset_real = pd. read_csv("p_real.txt", sep=" ", header=None, names=["V", "p"])
dataset_id2 = pd. read_csv("p_id2.txt", sep=" ", header=None, names=["V", "p"])
dataset_real2 = pd. read_csv("p_real2.txt", sep=" ", header=None, names=["V", "p"])
dataset_id3 = pd. read_csv("p_id3.txt", sep=" ", header=None, names=["V", "p"])
dataset_real3 = pd. read_csv("p_real3.txt", sep=" ", header=None, names=["V", "p"])

dataset_real12 = pd. read_csv("p_real12.txt", sep=" ", header=None, names=["V", "p"])
dataset_real22 = pd. read_csv("p_real22.txt", sep=" ", header=None, names=["V", "p"])
dataset_real32 = pd. read_csv("p_real32.txt", sep=" ", header=None, names=["V", "p"])
```

```python
dataset_real3
```

```python
V_id, p_id = dataset_id["V"], dataset_id["p"]
V_real, p_real = dataset_real["V"], dataset_real["p"]
V_id2, p_id2 = dataset_id2["V"], dataset_id2["p"]
V_real2, p_real2 = dataset_real2["V"], dataset_real2["p"]
V_id3, p_id3 = dataset_id3["V"], dataset_id3["p"]
V_real3, p_real3 = dataset_real3["V"], dataset_real3["p"]

V_real12, p_real12 = dataset_real12["V"], dataset_real12["p"]
V_real22, p_real22 = dataset_real22["V"], dataset_real22["p"]
V_real32, p_real32 = dataset_real32["V"], dataset_real32["p"]
```

```python
p_real3
```

```python
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(10)


plt.plot(V_id2[2:200], p_id2[2:200], label = "ideal gas")
plt.plot(V_real2[2:200], p_real2[2:200], label = "real gas, 4 comp", ls = ":")
plt.plot(V_real22[2:200], p_real22[2:200], label = "real gas, 2 comp", ls = ":")
plt.legend()
plt.ylabel("p, atm")
plt.xlabel("V, $m^3$")
plt.title("T = 100 K")
plt.grid(color = 'brown', lw = 0.3)
```

```python
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(10)


plt.plot(V_id3[2:200], p_id3[2:200], label = "ideal gas")
plt.plot(V_real3[2:200], p_real3[2:200], label = "real gas, 4 comp", ls = ":")
plt.plot(V_real32[2:200], p_real32[2:200], label = "real gas, 2 comp", ls = ":")
plt.legend()
plt.title("T = 200 K")
plt.ylabel("p, atm")
plt.xlabel("V, $m^3$")
plt.grid(color = 'brown', lw = 0.3)
```

```python
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(10)


plt.plot(V_id[2:200], p_id[2:200], label = "ideal gas")
plt.plot(V_real[2:200], p_real[2:200], label = "real gas, 4 comp", ls = ":")
plt.plot(V_real12[2:200], p_real12[2:200], label = "real gas, 2 comp", ls = ":")
plt.legend()
plt.title("T = 300 K")
plt.ylabel("p, atm")
plt.xlabel("V, $m^3$")
plt.grid(color = 'brown', lw = 0.3)
```

```python
plt.plot(V_id[10:100] - V_real[10:100], p_id[10:100] - p_real[10:100], label = "error")
plt.legend()
plt.xlabel("p, atm")
plt.ylabel("V, $10^4 \quad m^3$")
plt.grid(color = 'brown', lw = 0.3)
```
